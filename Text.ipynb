Research design and analysis framework
This project examines why the obesity rate in the United States remain high level around the world, and how this issue manifests across different U.S. states. Obesity is not a problem determined solely by individual behavior, but is closely related to the broader social, economic and policy contexts. Therefore, it is difficult to fully understand this phenomenon by relying only on single-level data.
Based on this consideration, this study adopts the idea of multi-level analytical framework. The analysis starts from the global level to provide an international comparative background for the obesity problem in the United States; then turn to the United States to examine the spatial distribution differences in state-level obesity rates; finally, after identifying states with high obesity rates, introduce socioeconomic and political variables to further exploratory analysis of these states. This step-by-step progression from broad context to more focused analysis helps establish background before attempting explanation.

Limitations of methods and data
When interpreting the analysis results, it is necessary to be cautious about the limitations of methods and data. First of all, all analysis is based on aggregate data at the state or county level, so there is a possibility of ecological fallacy. The statistical relationship at the group level does not necessarily reflect the behavioral mechanism at the individual level.
Secondly, correlation analysis itself cannot distinguish the causal direction. The relationship between obesity rate and socio-economic variables may be two-way, and may also be affected by other factors that are not included. In addition, political variables have been simplified in the quantitative process, making it difficult to fully reflect the actual complexity of state-level public health policies.
Despite these limitations, this study still provides a valuable exploratory perspective for understanding the spatial differences in obesity rates in the United States and their potential social background factors through the combination of multi-level data and multiple analytical methods.

Literature Review
Obesity has become one of the most urgent public health challenges in the United States. According to the data of the Behavioral Risk Factor Monitoring System (BRFSS), the obesity rate of American adults reached 42.4% in 2024 (Stierman et al., 2021). The southern and Appalachian states - especially West Virginia, Mississippi, Louisiana, Kentucky and Oklahoma - remain among the regions with the highest obesity rates in the United States. This trend shows that obesity should be studied within the framework of the social determinants of obesity.

The framework for healthy social determinants is shaped by Marmot (2005) and tCommission on Social Determinants of Health (2008), not just individual choices. This shift in theory is of great significance to the understanding of obesity because it shifts the focus from individual responsibility to structural inequality. Under this framework, socio-economic status, education level and political environment are key determinants of diet patterns, physical activity, health care access and final weight.

The relationship between economic conditions and obesity has been widely confirmed. Drewnowski & Specter (2004) put forward the poverty-obesity paradox of developed countries: contrary to intuition, poverty in rich countries does not lead to malnutrition. Another research also shows that the unit calorie cost of fresh agricultural products is 3-5 times that of processed food (Darmon & Drewnowski, 2008), and low-income communities are often food deserts that lack affordable health food retailers (Walker et al., 2010), and the long-term pressure brought about by economic insecurity will disrupt metabolic regulation (Offer et al., 2010). The meta-analysis of 54 studies conducted by McLaren (2007) confirmed that there was a significant negative correlation between the socioeconomic status and obesity rate in developed countries, with an average correlation coefficient of about -0.45.

As another core social determinant, the influence mechanism of education is more complex. Cutler and Lleras-Muney (2010) summarised the impact of education on health through three ways in a review published in the journal Economic Outlook: productivity effect (education enhances the ability to understand and apply health information), resource allocation effect (through the impact Occupation and income indirectly affect access to health resources) and time preference effect (education affects people's attention to future returns). A 25-year follow-up study conducted by Cohen et al. (2013) found that the risk of obesity is reduced by about 7% for every year of education, and this effect is still significant after controlling the income factor, indicating that education has a health protection effect independent of economic status. Grossman (2006) believes that education makes individuals more efficient "healthy producers".

In recent years, political factors have received increasing attention as an emerging dimension of health determinants. Beckfield & Krig (2009) proposed the concept of "political determinants" to empharise how policy choices and power distributions systematically affect health outcomes. A study published by Montez et al. (2017) in the American Journal of Public Health quantified policy differences between U.S. states and found that policy variables could explain up to 50% of interstate life expectancy differences. However, a study published by Wasfy et al. (2017) in the Journal of the American Medical Association found that the relationship between political tendencies and health outcomes is very complex: counties with high Trump support in 2016 usually had higher mortality rates. However, the study of Gellman et al. (2024) shows that due to the "McPerson's paradox" and demographic effects, the political and health relationship at different geographical levels may be completely opposite.

Although the existing literature on social determinants and obesity is quite rich, there is still a lack of systematic comparative research on the high obesity rate in specific regions, and the interaction between economic, educational and political factors is still unclear. This study aims to test the applicability of the social determinant theory in these "extreme cases" by analyzing the county-level data of five states with high obesity rates in the United States.

Discussion
The empirical analysis of this study not only confirms the core predictions of the literature, but also shows important regional differences and theoretical contradictions. Next, it will be explained one by one from the three perspectives of economy, education and politics.

The most important finding of the study is that there is a significant negative correlation between the medan household income and the adult obesity rate. This conclusion was verified in all five states analyzed: Mississippi (r = -0.752, n=83), Kentucky (r = -0.689, n=121), Louisiana (r = -0.683, n=65), West Virginia (r = - 0.552, n=56) and Oklahoma (r = -0.510, n=78). This finding is highly consistent with the "poverty-obesity paradox" proposed by Drewnowski and Spectre (2004), which directly supports the mechanism by which economic poverty leads to obesity by restricting people's access to healthy food.

However, the correlation intensity observed in the data analysis results is generally higher than the average (r ≈ -0.45) reported in the McLaren (2007) meta-analysis. In particular, Mississippi's -0.752 has almost reached the high limit of strong correlation. This difference can be understood from two aspects: first, the impact of extreme cases - the state examined in this study is one of the poorest states in the United States, representing a typical case of "extreme poverty - extreme obesity". As Offer et al. (2010) pointed out, the marginal effect of income will be amplified at the level of absolute poverty. As one of the poorest states in the United States, Mississippi has a county-level income of between $30,000 and $70,000 and an obesity rate of more than 32% to 50%, which makes the income effect more significant. Secondly, the advantage of county-level analysis is that county-level data can capture the variability of community factors (such as the local food environment), which are difficult to measure in individual or state-level data (Walker et al., 2010).

All states have confirmed a negative correlation between education level and obesity rate, ranging from -0.429 (West Virginia) to -0.655 (Louisiana). This is consistent with the results of a longitudinal study by Cohen et al. (2013), which found that the risk of obesity is reduced by 7% for every year of education. Comparing our data with Cohen's research results, the obesity rate in Louisiana has decreased by about 8 percentage points as the proportion of the population with a college degree increased from 30% to 70%, which is similar.

The differences in educational effects between states are worthy of attention. The significant educational effect of Louisiana (r = -0.655) may reflect the mechanism of "healthy knowledge production" empharised by Kenkel (1991): in areas with highly differentiated educational resources, the gap in health literacy between educated and uneducated people is greater. On the contrary, the weaker educational effect of West Virginia (r = -0.429) echoes Greenhalgh's (2015) criticism that the marginal effect of education will weaken in areas facing serious structural obstacles to development (such as the economic collapse caused by deindustrialization).

The most unexpected finding of this study is that there is an inverse relationship between political tendencies and obesity rates in different states, which is the opposite of the national model reported by Vasfi et al. (2017). In Kentucky (r = +0.451) and West Virginia (r = +0.498), Republican support is positively correlated with obesity, which is consistent with Wasfy's research results. However, Mississippi showed a strong negative correlation (r = -0.789), which means that the higher the Republican support, the lower the obesity rate, which is exactly the opposite of expectations.

Given the warning of the Simpson paradox by Gelman et al. (2008), this anomaly is understandable. Mississippi's racial political structure is unique: African-Americans make up 38% of the state's population (the highest in the United States), and race and politics almost completely overlap - the majority of African-American communities support the Democratic Party, while the white community is highly Republican. At the same time, data from the U.S. Centers for Disease Control and Prevention shows that the obesity rate of African Americans (49.6%) is significantly higher than that of whites (42.2%), which stems from historical apartheid and economic discrimination (Williams & Mohammed, 2013). Therefore, at the county level, the obesity rate in counties with high Democratic support (mainly composed of African-American residents) is also high, and there is a statistical negative correlation between the two. It should be noted that this finding reveals a blind spot in the theory of political determinants proposed by Bakerfield and Krieger (2009): using election results to represent the political environment ignores the impact of demographic factors.

Reflection
Although this study has achieved remarkable empirical results, there are still many methodological and theoretical limitations, which are worth further discussion.

The most fundamental limitation lies in the problem of causal inference. The correlation in cross-sectional data is a mixture of causality, inverse causality and common cause. Pearl and Mackenzie (2018) empharised that observational data can only reveal correlation, not causality. Although in theory, income affects food choices, which in turn affects obesity, there may also be a reverse causal relationship: obesity will lead to employment discrimination, thus reducing income. To establish a causal relationship, randomised controlled trials or quasi-experimental design are required, and these methods are often difficult to implement in large-scale population studies.

In addition, the county-level summary data may cover up the heterogeneity in the county. Robinson's classic paper (1950) points out that the relevance of the group level cannot be extraported to the individual level. The negative correlation between politics and obesity in Mississippi is a classic case, reflecting the demographic effect rather than the causal relationship at the individual level. Therefore, the aggregate data used in this study may underestimate the real impact of socio-economic status at the individual level. Future research should adopt a multi-layer model and analyze the individual and community levels at the same time.

In terms of methodological design, this study adopts a single linear research model, which fails to measure key intermediary paths such as supermarket density and working hours, so it fails to determine the most effective intervention entry point. In addition, this study ignores potential regulatory variables such as community evaluation and local public health inputs, although these variables can explain the differences in obesity rates between counties with similar socioeconomic conditions.

Finally, from a policy perspective, public health interventions often have the color of patriarchal style and ethical judgment (Leichter, 2003). The results of the study may still be misused; for example, the high obesity rate in low-education groups may be interpreted as a lack of self-discipline. Therefore, policies should be based on collective responsibility rather than individual blame, and create a social environment where everyone can flourish. In addition, physiological diversity should be recognised to avoid stigmatising the research results.
